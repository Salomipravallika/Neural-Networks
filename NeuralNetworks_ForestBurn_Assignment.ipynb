##### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import keras
from sklearn.preprocessing import StandardScaler
```

##### Loading dataset

```python
df1=pd.read_csv(r'C:\Users\hp\Downloads\forestfires.csv')
df1.head()
```

```python
df2=df1.drop(['month','day'],axis=1)
df2.head()
```

##### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
```

```python
#Applying Label encoding to size_category column and adding result in new column named y
df2['size_category']=lb.fit_transform(df2['size_category'])
df2
```

```python
df=df2.values
df.shape
```

```python
x=df[:,0:28]
y=df[:,-1]
x.shape
```

##### Importing libraries for creating model

```python
from sklearn.model_selection import GridSearchCV, KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import Adam
```

##### Creating Model

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=28, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
model.fit(x, y, validation_split=0.33,epochs=100, batch_size=5)
```

##### Accuracy

```python
# evaluate the model
scores = model.evaluate(x,y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100)) 
```

##### Inference
- The model is a good predictor for the burned area of Forest Fires
